## Implementação de Classes

````python
class Stack:
    # Construtor
    def __init__(self):
        self.items = []
    # Métodos
    def isEmpty(self):
        return self.items == []
    def push(self, item):
        self.items.append(item)
    def pop(self):
        if self.items:
            return self.items.pop()
    def peek(self):
        if self.items: 
            return self.items[-1]
    def size(self):
        return len(self.items)

class Queue:
    # Construtor
    def __init__(self):
        self.items = []
    # Métodos
    def isEmpty(self):
        return self.items == []
    def enqueue(self, item):
        self.items.insert(0,item)
    def dequeue(self):
        return self.items.pop()
    def start(self):
        return self.items[-1]
    def size(self):
        return len(self.items)  
````

### Questão 01 - Dequeue de Caracteres    
Há muitas maneiras divertidas de se fornecer e executar comandos, por exemplo com uma abordagem **filacetamol** para processar os caracteres.É muito simples, uma letra significa a operação **enqueue** e um asterisco significa a operação **dequeue**.Eis um exemplo de uma instrução em **filacetamol**(na primeira linha) e o resultado do processamento dela (na segunda).

```bash
>>> ola*** mundo
ola
```

Imprima a sequência de caracteres retornados pelas operações de **dequeue** quando esta sequência de operações é realizada em uma fila de caracteres inicialmente vazia.

**Entrada**  
Sequência de caracteres em uma linha.

**Saída**  
Sequência de caracteres processados pelas operações **dequeue**.

**Por exemplo:**
| **Input**                                  | **Resultado**              |
|:------------------------------------------|:---------------------------|
| ola\*\*\* mundo                            | ola                         |
| c > python\*\*\*                           | c >                         |
| rir\*\*\*,\* o brev\*\*\*\*\*\*\*e verb\*\*\*\*\*\*o \*rir\*\*\*\* | rir, o breve verbo rir  |

````python
def fila_de_caracteres(frase):
    q = Queue()
    s = ""
    for caracter in frase:
          if caracter != '*':
                q.enqueue(caracter)
          else:
                # Concatena as caracteres para mostrá-los 
                s += q.dequeue()  
    return s

frase = input()
print(fila_de_caracteres(frase))
````

### Questão 06 - Drummondichlorian!
O Mestre Jedi D'Costa desenvolveu uma nova estrutura de dados, a **drummondichlorian**, que tem propriedades ainda desconhecidas. Sabe-se que, entre outras coisas, ela pode atuar com comportamentos **FIFO** ou **LIFO**. Às vezes, ela também serve para priorizar a Força dos elementos, apresentando o comportamento de uma **fila de prioridade**, onde o elemento de maior valor é o de maior prioridade.

Como parte de seu treinamento, você deve identificar o comportamento dessa nova ferramenta!  
Dessa forma, dada a sequência de instruções de inserção/remoção em uma drummondichlorian, indique com qual estrutura de dados tradicional ela se parece.

**Entrada**  
A entrada contém **5 casos de teste**.  
Para cada caso:
- A primeira linha apresenta um inteiro 𝑁 (0 ≤ 𝑁 ≤ 100) indicando a quantidade de instruções realizadas.

  - `in X` — inserção do valor X.
  - `out X` — remoção do valor X.

Considere que 0 ≤ X ≤ 1000.

**Saída**  
Para cada caso de teste, analise o comportamento da drummondichlorian e diga com qual estrutura de dados ela se parece. As opções são:
- `"LIFO"` — se ela se comportar como uma pilha.
- `"FIFO"` — se ela se comportar como uma fila.
- `"AIPO"` — se parecer com uma fila de prioridades ("any in, priority out").
- `"no hay!"` — se o comportamento não corresponder a nenhuma estrutura conhecida.
- `"uai?"` — se houver mais de uma possibilidade.

**Observação**  
O Mestre D'Costa garante que:
- Nunca será inserido algo que não seja um número.
- Nunca haverá uma remoção em uma drummondichlorian vazia.

**Exemplificação**:  
No primeiro caso de teste do primeiro exemplo:
- Inserções: 17, depois 13.
- Remoção: 17.
- Não é pilha (LIFO), mas também não é possível determinar se é fila (FIFO) ou fila de prioridade (AIPO).

No segundo caso:
- Inserções: 1, 2, 3.
- Remoções: 1, 2 (na ordem de inserção).
- É uma fila (FIFO).

**Por exemplos:**
| **Input** | **Resultado** |
|:---|:---|
| 3<br>in 17<br>in 13<br>out 17<br>5<br>in 1<br>in 2<br>in 3<br>out 1<br>out 2<br>5<br>in 23<br>in 30<br>out 30<br>in 1<br>out 23<br>6<br>in 100<br>in 90<br>in 80<br>out 80<br>out 90<br>out 100<br>5<br>in 1<br>in 2<br>out 1<br>in 1<br>out 1 | uai?<br>FIFO<br>AIPO<br>LIFO<br>no hay! |
| 4<br>in 11<br>in 2<br>out 2<br>out 11<br>4<br>in 1<br>out 1<br>in 1<br>out 1<br>1<br>in 42<br>6<br>in 1<br>in 2<br>in 3<br>out 3<br>out 2<br>out 1<br>5<br>in 1<br>in 2<br>out 2<br>in 3<br>out 1 | LIFO<br>uai?<br>uai?<br>uai?<br>no hay! |
| 5<br>in 22<br>in 222<br>in 2<br>out 22<br>out 222<br>4<br>in 4<br>in 5<br>in 6<br>out 4<br>4<br>in 4<br>in 5<br>in 6<br>out 6<br>4<br>in 6<br>in 5<br>in 4<br>out 6<br>4<br>in 5<br>in 6<br>in 4<br>out 6 | FIFO <br> FIFO <br> uai? <br> uai? <br> AIPO |


````python
for _ in range(5):
    qtde_instrucoes = int(input())
    instrucoes = []

    # Armanezar as instruções
    for _ in range(qtde_instrucoes):
        instrucao = input().strip()
        partes = instrucao.split()
        opcao = partes[0]
        X = int(partes[1])
        instrucoes.append((opcao, X))

    # Flags
    possivel_pilha = True
    possivel_fila = True
    possivel_fila_com_prioridade = True

    # Pilha (LIFO)
    pilha = []
    for opcao, X in instrucoes:
        if opcao == 'in':
            pilha.append(X)
        else:  # opcao == 'out'
            if not pilha or pilha[-1] != X:
                possivel_pilha = False
            else:
                pilha.pop()

    # Fila (FIFO)
    fila = []
    for opcao, X in instrucoes:
        if opcao == 'in':
            fila.append(X)
        else:  # opcao == 'out'
            if not fila or fila[0] != X:
                possivel_fila = False
            else:
                fila.pop(0)

    # Fila de Prioridade (AIPO) --> Prioridade = maior valor
    fila_prioridade = []
    for opcao, X in instrucoes:
        if opcao == 'in':
            fila_prioridade.append(X)
        else:  # opcao == 'out'
            if not fila_prioridade or max(fila_prioridade) != X:
                possivel_fila_com_prioridade = False
            else:
                fila_prioridade.remove(X)

    # Resultado
    possibilidades = []
    if possivel_fila:
        possibilidades.append("FIFO")
    if possivel_pilha:
        possibilidades.append("LIFO")
    if possivel_fila_com_prioridade:
        possibilidades.append("AIPO")

    if len(possibilidades) > 1:
        print("uai?")
    elif len(possibilidades) == 1:
        print(possibilidades[0])
    else:
        print("no hay!")
````

### Questão 08 - Posto de Pesagem
Na última estação chuvosa, uma rodovia do Distrito Federal ficou muito esburacada, sendo motivo de reclamação de diversos condutores que transitam nela diariamente.  
Ações de fiscalização por parte dos agentes rodoviários verificaram que vários caminhões transitavam acima do peso permitido para a rodovia, o que provocou rachaduras que pioraram com as chuvas.

Tentando resolver o problema, os órgãos de trânsito competentes instalaram um **posto de pesagem** com o objetivo de fiscalizar se os pesos dos veículos ultrapassam um limite de peso `P` (em kg).

Os fiscais deste posto de pesagem sabem que é impossível abordar todos os veículos, então consideram um **fator `F`** que denota uma amostragem dos veículos a serem pesados:
- Se `F = 1`, todos os veículos são fiscalizados.
- Se `F = 2`, os veículos são fiscalizados alternadamente (um fiscalizado, um liberado).
- Se `F = 3`, a cada três veículos, um é fiscalizado, dois são liberados, e assim sucessivamente.

Alguns caminhoneiros "fora da lei" tentam passar sem serem abordados, mas:
- Se forem abordados e estiverem acima do limite, são barrados.
- Para contornar a situação, descartam `2 kg` da carga total e retornam ao final da fila, tentando novamente.

**Tempos** ($t_i$) gastos no posto de pesagem:
- Um veículo **não abordado**: $t_i = 1$ minuto.
- Um veículo **abordado e dentro do limite**: $t_i = 5$ minutos.
- Um veículo **abordado e acima do limite**: $t_i = 10$ minutos (e retorna para o fim da fila).

**Entrada**  
A primeira linha da entrada contém três números inteiros separados por espaço:
- `N` — número de veículos (1 ≤ N ≤ $10^4$)
- `F` — fator de amostragem (1 ≤ F ≤ 100)
- `P` — limite de peso da rodovia (1 ≤ P ≤ $10^3$)

A segunda linha contém `N` valores inteiros $a_1,...,a_n$ (1 ≤ $a_i$ ≤ $10^3$) representando o peso de cada veículo.

**Saída**  
Imprima um valor inteiro correspondente ao **tempo total** (em minutos) para que todos os veículos passem pelo posto de pesagem.

**Observações**  
- No primeiro exemplo, o limite é `P = 7` e o fator é `F = 1`, ou seja, todos os veículos são fiscalizados.
- Os veículos 1 a 4 estão dentro do limite, gastando `5` minutos cada.
- O veículo 5 está acima do limite (9 kg > 7 kg), então:
  - É fiscalizado (gasta 10 minutos), descarta 2 kg (fica com 7 kg) e volta ao final da fila.
  - É fiscalizado novamente e liberado (5 minutos).

Tempo total: `5 + 5 + 5 + 5 + 10 + 5 = 35` minutos.

**Por exemplos**
| **Input** | **Resultado** |
|:---|:---|
| 5 1 7<br>4 2 6 3 9 | 35 |
| 2 1 5<br>7 4 | 20 |
| 4 2 8<br>2 6 1 3 | 12 |

````python
entrada = input()
partes = entrada.split()
qtde_caminhoes = int(partes[0])
fator_amostragem = int(partes[1])
limite_peso = int(partes[2])

pesos = input()
pesos_caminhoes = pesos.split()

q = Queue()
for i in pesos_caminhoes:
    q.enqueue(int(i))

tempo_total = 0
posicao_caminhao = 1

while not q.isEmpty():
     # Verifica se o caminhão na posição atual precisa ser fiscalizado
     if ((posicao_caminhao - 1) % fator_amostragem) == 0:
        # Caminhão --> fiscalizado e volta ao final da fila
        if q.start() > limite_peso:
            q.enqueue(q.dequeue() - 2)
            tempo_total += 10
        else:
            # Caminhão --> Fiscalizado e liberado
            tempo_total += 5
            q.dequeue()
     else:
        # Caminhão não é fiscalizado
        tempo_total += 1
        q.dequeue()
     
     posicao_caminhao += 1

print(tempo_total)
````

| `Carlos Gabriel Joao Vitor`<br>`1` | Pessoa da frente: Gabriel<br>Pessoa do fim: Carlos |
| `Carlos Gabriel Joao Vitor`<br>`4` | Pessoa da frente: Carlos<br>Pessoa do fim: Vitor |

````python
nomes = input()
x = int(input())

nome = nomes.split()
q = Queue()

for i in range(len(nome)):
    q.enqueue(nome[i])

numero_avancos = 0
while x > numero_avancos:
    q.enqueue(q.dequeue())
    numero_avancos += 1

primeira_pessoa = q.start()

s = Stack()
while not q.isEmpty():
    s.push(q.dequeue())

ultima_pessoa = s.peek()

print("Pessoa da frente:", primeira_pessoa + "\nPessoa do fim:", ultima_pessoa)
````

### Questão 10 - Remoção de Duplicatas
Na matemática, o parêntese representa a prioridade de resolução em uma expressão algébrica. Por exemplo, existe uma diferença entre:

- \( 8/(2*(2+2)) = 1 \)
- \( (8/2)*(2+2) = 16 \)

Além dos parênteses, apesar de menos usuais, os símbolos de colchetes e chaves também podem ser utilizados para criar uma ordem de prioridade na resolução, onde a prioridade é:

- `( )` - Parêntese
- `[ ]` - Colchete
- `{ }` - Chave

Para todos esses casos descritos, um fator em comum é que, se qualquer uma dessas representações de prioridades for duplicada em uma expressão, ela é insignificante.  
Exemplo:

- \( 5*((1001+110)) \) é equivalente a \( 5*(1001+110) \).

Sua tarefa é, dada uma lista de expressões algébricas válidas, verificar se há algum símbolo de prioridade duplicado dentro de cada expressão.

**Entrada**  
A entrada é composta por várias linhas.  
A primeira linha possui um inteiro  N $(1 \leq N \leq 100)$ que indica o número de expressões a serem verificadas.  
A seguir, as próximas \( N \) linhas possuem, cada uma, uma string \( E \) (de comprimento entre 4 e 1000, inclusive), sem espaços, que representa a expressão algébrica a ser analisada.

**Saída**  
Para cada expressão, a saída deve ser:

- `"A expressão possui duplicata."` se houver um símbolo de prioridade duplicado dentro da expressão, ou
- `"A expressão não possui duplicata."` caso contrário,

de acordo com os exemplos abaixo.

| Input | Resultado |
|:-----:|:---------:|
| 1 <br> 4-77*21 | A expressão não possui duplicata. |
| 1 <br> ((4-77))*21 | A expressão possui duplicata. |
| 3 <br> [(4-77)]*21 <br> ((20+20))-96 <br> 10*20 | A expressão não possui duplicata.<br> A expressão possui duplicata.<br> A expressão não possui duplicata. |

````python
def duplicata(sim_abre,simb_fecha,expressao):
    s = Stack()
    for c in expressao:
        if c == simb_fecha:
            if s.peek() == sim_abre:
                return True
            else:
                while s.peek() != sim_abre:
                    s.pop()
                s.pop()
        else:
            s.push(c)
    return False

N = int(input()) 
for _ in range(N):
    expressao = input()

    if duplicata('(',')',expressao) or duplicata('[',']',expressao) or duplicata('{','}',expressao):
        print('A expressão possui duplicata.')
    else:
        print('A expressão não possui duplicata.')
````
