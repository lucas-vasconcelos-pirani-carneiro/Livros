## Implementa√ß√£o de Classes

````python
class Stack:
    # Construtor
    def __init__(self):
        self.items = []
    # M√©todos
    def isEmpty(self):
        return self.items == []
    def push(self, item):
        self.items.append(item)
    def pop(self):
        if self.items:
            return self.items.pop()
    def peek(self):
        if self.items: 
            return self.items[-1]
    def size(self):
        return len(self.items)

class Queue:
    # Construtor
    def __init__(self):
        self.items = []
    # M√©todos
    def isEmpty(self):
        return self.items == []
    def enqueue(self, item):
        self.items.insert(0,item)
    def dequeue(self):
        return self.items.pop()
    def start(self):
        return self.items[-1]
    def size(self):
        return len(self.items)  
````

### Quest√£o 01 - Dequeue de Caracteres    
H√° muitas maneiras divertidas de se fornecer e executar comandos, por exemplo com uma abordagem **filacetamol** para processar os caracteres.√â muito simples, uma letra significa a opera√ß√£o **enqueue** e um asterisco significa a opera√ß√£o **dequeue**.Eis um exemplo de uma instru√ß√£o em **filacetamol**(na primeira linha) e o resultado do processamento dela (na segunda).

```bash

```

Imprima a sequ√™ncia de caracteres retornados pelas opera√ß√µes de **dequeue** quando esta sequ√™ncia de opera√ß√µes √© realizada em uma fila de caracteres inicialmente vazia.

**Entrada**  
Sequ√™ncia de caracteres em uma linha.

**Sa√≠da**  
Sequ√™ncia de caracteres processados pelas opera√ß√µes **dequeue**.

**Por exemplo:**
| **Input**                                  | **Resultado**              |
|:------------------------------------------|:---------------------------|
| ola\*\*\* mundo                            | ola                         |
| c > python\*\*\*                           | c >                         |
| rir\*\*\*,\* o brev\*\*\*\*\*\*\*e verb\*\*\*\*\*\*o \*rir\*\*\*\* | rir, o breve verbo rir  |

````python
def fila_de_caracteres(frase):
    q = Queue()
    s = ""
    for caracter in frase:
          if caracter != '*':
                q.enqueue(caracter)
          else:
                # Concatena as caracteres para mostr√°-los 
                s += q.dequeue()  
    return s

frase = input()
print(fila_de_caracteres(frase))
````

### Quest√£o 03 - Pop de Caracteres
H√° muitas maneiras divertidas de se fornecer e executar comandos, por exemplo com uma abordagem **pilhacetamol** para processar os caracteres.√â muito simples, uma letra significa a opera√ß√£o **push** e um asterisco significa a opera√ß√£o **pop**.Eis um exemplo de uma instru√ß√£o em **pilhacetamol** (na primeira linha) e o resultado do processamento dela (na segunda).

Imprima a sequ√™ncia de caracteres retornados pelas opera√ß√µes de dequeue quando esta sequ√™ncia de opera√ß√µes √© realizada em uma fila de caracteres inicialmente vazia.

**Entrada**  
Sequ√™ncia de caracteres em uma linha.

**Sa√≠da**  
Sequ√™ncia de caracteres processados pelas opera√ß√µes **pop**.

**Por exemplo:**
| **Input**                                  | **Resultado**              |
|:------------------------------------------|:---------------------------|
| ola\*\*\* mundo                            | alo                         |
| nohtyp > c\*\*\*\*\*\*\*\*\*\*                           | c > python                         |
| rir\*\*\*,\* o brev\*\*\*\*\*\*\*e verb\*\*\*\*\*\*o \*rir\*\*\*\* | rir,verb o brev e rir  |

````python
def inverter_frase(frase):
     s = Stack()
     frase_invertida = ""
     for caracter in frase:
          if caracter != "*":
               s.push(caracter)
          else:
               frase_invertida += s.pop()
     return frase_invertida

frase = input()
print(inverter_frase(frase))
````

````python
s = Stack()

while True:
    peso = int(input())
    if peso != 0:
        s.push(peso)
    else:
          break
    
peso_desejado = int(input())

soma_pesos = 0
qtde_anilhas = 0

while True:
    if not s.items:  # Se a pilha estiver vazia, pare
        break
    topo = s.peek()
    print('Peso retirado:', topo)
    soma_pesos += s.pop()
    qtde_anilhas += 1
    if topo == peso_desejado:
        break

print("Anilhas retiradas:", qtde_anilhas)
print("Peso total movimentado:", soma_pesos)
````

### Quest√£o 06 - Drummondichlorian!
O Mestre Jedi D'Costa desenvolveu uma nova estrutura de dados, a **drummondichlorian**, que tem propriedades ainda desconhecidas. Sabe-se que, entre outras coisas, ela pode atuar com comportamentos **FIFO** ou **LIFO**. √Äs vezes, ela tamb√©m serve para priorizar a For√ßa dos elementos, apresentando o comportamento de uma **fila de prioridade**, onde o elemento de maior valor √© o de maior prioridade.

Dessa forma, dada a sequ√™ncia de instru√ß√µes de inser√ß√£o/remo√ß√£o em uma drummondichlorian, indique com qual estrutura de dados tradicional ela se parece.

**Entrada**

Para cada caso:
- A primeira linha apresenta um inteiro ùëÅ (0 ‚â§ ùëÅ ‚â§ 100) indicando a quantidade de instru√ß√µes realizadas.

Considere que 0 ‚â§ X ‚â§ 1000.

**Sa√≠da**  
Para cada caso de teste, analise o comportamento da drummondichlorian e diga com qual estrutura de dados ela se parece. As op√ß√µes s√£o:
- `"LIFO"` ‚Äî se ela se comportar como uma pilha.
- `"FIFO"` ‚Äî se ela se comportar como uma fila.
- `"AIPO"` ‚Äî se parecer com uma fila de prioridades ("any in, priority out").
- `"no hay!"` ‚Äî se o comportamento n√£o corresponder a nenhuma estrutura conhecida.
- `"uai?"` ‚Äî se houver mais de uma possibilidade.

**Observa√ß√£o**  
O Mestre D'Costa garante que:
- Nunca ser√° inserido algo que n√£o seja um n√∫mero.
- Nunca haver√° uma remo√ß√£o em uma drummondichlorian vazia.

**Exemplifica√ß√£o**:  
No primeiro caso de teste do primeiro exemplo:
- Inser√ß√µes: 17, depois 13.
- Remo√ß√£o: 17.
- N√£o √© pilha (LIFO), mas tamb√©m n√£o √© poss√≠vel determinar se √© fila (FIFO) ou fila de prioridade (AIPO).

No segundo caso:
- Inser√ß√µes: 1, 2, 3.
- Remo√ß√µes: 1, 2 (na ordem de inser√ß√£o).
- √â uma fila (FIFO).

**Por exemplos:**
| **Input** | **Resultado** |
|:---|:---|
| 3<br>in 17<br>in 13<br>out 17<br>5<br>in 1<br>in 2<br>in 3<br>out 1<br>out 2<br>5<br>in 23<br>in 30<br>out 30<br>in 1<br>out 23<br>6<br>in 100<br>in 90<br>in 80<br>out 80<br>out 90<br>out 100<br>5<br>in 1<br>in 2<br>out 1<br>in 1<br>out 1 | uai?<br>FIFO<br>AIPO<br>LIFO<br>no hay! |
| 4<br>in 11<br>in 2<br>out 2<br>out 11<br>4<br>in 1<br>out 1<br>in 1<br>out 1<br>1<br>in 42<br>6<br>in 1<br>in 2<br>in 3<br>out 3<br>out 2<br>out 1<br>5<br>in 1<br>in 2<br>out 2<br>in 3<br>out 1 | LIFO<br>uai?<br>uai?<br>uai?<br>no hay! |
| 5<br>in 22<br>in 222<br>in 2<br>out 22<br>out 222<br>4<br>in 4<br>in 5<br>in 6<br>out 4<br>4<br>in 4<br>in 5<br>in 6<br>out 6<br>4<br>in 6<br>in 5<br>in 4<br>out 6<br>4<br>in 5<br>in 6<br>in 4<br>out 6 | FIFO <br> FIFO <br> uai? <br> uai? <br> AIPO |


````python
for _ in range(5):
    qtde_instrucoes = int(input())
    instrucoes = []

    # Armanezar as instru√ß√µes
    for _ in range(qtde_instrucoes):
        instrucao = input().strip()
        partes = instrucao.split()
        opcao = partes[0]
        X = int(partes[1])
        instrucoes.append((opcao, X))

    # Flags
    possivel_pilha = True
    possivel_fila = True
    possivel_fila_com_prioridade = True

    # Pilha (LIFO)
    pilha = []
    for opcao, X in instrucoes:
        if opcao == 'in':
            pilha.append(X)
        else:  # opcao == 'out'
            if not pilha or pilha[-1] != X:
                possivel_pilha = False
            else:
                pilha.pop()

    # Fila (FIFO)
    fila = []
    for opcao, X in instrucoes:
        if opcao == 'in':
            fila.append(X)
        else:  # opcao == 'out'
            if not fila or fila[0] != X:
                possivel_fila = False
            else:
                fila.pop(0)

    # Fila de Prioridade (AIPO) --> Prioridade = maior valor
    fila_prioridade = []
    for opcao, X in instrucoes:
        if opcao == 'in':
            fila_prioridade.append(X)
        else:  # opcao == 'out'
            if not fila_prioridade or max(fila_prioridade) != X:
                possivel_fila_com_prioridade = False
            else:
                fila_prioridade.remove(X)

    # Resultado
    possibilidades = []
    if possivel_fila:
        possibilidades.append("FIFO")
    if possivel_pilha:
        possibilidades.append("LIFO")
    if possivel_fila_com_prioridade:
        possibilidades.append("AIPO")

    if len(possibilidades) > 1:
        print("uai?")
    elif len(possibilidades) == 1:
        print(possibilidades[0])
    else:
        print("no hay!")
````

  - √â fiscalizado novamente e liberado (5 minutos).

Tempo total: `5 + 5 + 5 + 5 + 10 + 5 = 35` minutos.

**Por exemplos**
| **Input** | **Resultado** |
|:---|:---|
| 5 1 7<br>4 2 6 3 9 | 35 |
| 2 1 5<br>7 4 | 20 |
| 4 2 8<br>2 6 1 3 | 12 |

````python
entrada = input()
partes = entrada.split()
qtde_caminhoes = int(partes[0])
fator_amostragem = int(partes[1])
limite_peso = int(partes[2])

pesos = input()
pesos_caminhoes = pesos.split()

q = Queue()
for i in pesos_caminhoes:
    q.enqueue(int(i))

tempo_total = 0
posicao_caminhao = 1

while not q.isEmpty():
     # Verifica se o caminh√£o na posi√ß√£o atual precisa ser fiscalizado
     if ((posicao_caminhao - 1) % fator_amostragem) == 0:
        # Caminh√£o --> fiscalizado e volta ao final da fila
        if q.start() > limite_peso:
            q.enqueue(q.dequeue() - 2)
            tempo_total += 10
        else:
            # Caminh√£o --> Fiscalizado e liberado
            tempo_total += 5
            q.dequeue()
     else:
        # Caminh√£o n√£o √© fiscalizado
        tempo_total += 1
        q.dequeue()
     
     posicao_caminhao += 1

print(tempo_total)
````

Voc√™ foi contratado pela equipe de desenvolvedores chamada **P√≥s-vida**, do CEO Fagundes, para programar o comportamento do Inferno.

Dada a quantidade de "eternidades" passadas, indique as pessoas que est√£o no in√≠cio e no fim da fila.

**Sa√≠da**  
A sa√≠da consiste no nome das pessoas que est√£o:
- na frente da fila, e
- no fim da fila,depois de `X` avan√ßos.  

Apresente-os, nessa ordem, um por linha, conforme os exemplos abaixo:

| `Carlos Gabriel Joao Vitor`<br>`0` | Pessoa da frente: Carlos<br>Pessoa do fim: Vitor |

````python
nomes = input()
x = int(input())

nome = nomes.split()
q = Queue()

for i in range(len(nome)):
    q.enqueue(nome[i])

numero_avancos = 0
while x > numero_avancos:
    q.enqueue(q.dequeue())
    numero_avancos += 1

primeira_pessoa = q.start()

s = Stack()
while not q.isEmpty():
    s.push(q.dequeue())

ultima_pessoa = s.peek()

print("Pessoa da frente:", primeira_pessoa + "\nPessoa do fim:", ultima_pessoa)
````

### Quest√£o 10 - Remo√ß√£o de Duplicatas
Na matem√°tica, o par√™ntese representa a prioridade de resolu√ß√£o em uma express√£o alg√©brica. Por exemplo, existe uma diferen√ßa entre:

- \( 8/(2*(2+2)) = 1 \)
- \( (8/2)*(2+2) = 16 \)

Al√©m dos par√™nteses, apesar de menos usuais, os s√≠mbolos de colchetes e chaves tamb√©m podem ser utilizados para criar uma ordem de prioridade na resolu√ß√£o, onde a prioridade √©:

- `( )` - Par√™ntese
- `[ ]` - Colchete
- `{ }` - Chave

Para todos esses casos descritos, um fator em comum √© que, se qualquer uma dessas representa√ß√µes de prioridades for duplicada em uma express√£o, ela √© insignificante.  
Exemplo:

- \( 5*((1001+110)) \) √© equivalente a \( 5*(1001+110) \).

Sua tarefa √©, dada uma lista de express√µes alg√©bricas v√°lidas, verificar se h√° algum s√≠mbolo de prioridade duplicado dentro de cada express√£o.

**Entrada**  
A entrada √© composta por v√°rias linhas.  
A primeira linha possui um inteiro  N $(1 \leq N \leq 100)$ que indica o n√∫mero de express√µes a serem verificadas.  
A seguir, as pr√≥ximas \( N \) linhas possuem, cada uma, uma string \( E \) (de comprimento entre 4 e 1000, inclusive), sem espa√ßos, que representa a express√£o alg√©brica a ser analisada.

**Sa√≠da**  
Para cada express√£o, a sa√≠da deve ser:

- `"A express√£o possui duplicata."` se houver um s√≠mbolo de prioridade duplicado dentro da express√£o, ou
- `"A express√£o n√£o possui duplicata."` caso contr√°rio,

de acordo com os exemplos abaixo.

| Input | Resultado |
|:-----:|:---------:|
| 1 <br> 4-77*21 | A express√£o n√£o possui duplicata. |
| 1 <br> ((4-77))*21 | A express√£o possui duplicata. |
| 3 <br> [(4-77)]*21 <br> ((20+20))-96 <br> 10*20 | A express√£o n√£o possui duplicata.<br> A express√£o possui duplicata.<br> A express√£o n√£o possui duplicata. |

````python
def duplicata(sim_abre,simb_fecha,expressao):
    s = Stack()
    for c in expressao:
        if c == simb_fecha:
            if s.peek() == sim_abre:
                return True
            else:
                while s.peek() != sim_abre:
                    s.pop()
                s.pop()
        else:
            s.push(c)
    return False

N = int(input()) 
for _ in range(N):
    expressao = input()

    if duplicata('(',')',expressao) or duplicata('[',']',expressao) or duplicata('{','}',expressao):
        print('A express√£o possui duplicata.')
    else:
        print('A express√£o n√£o possui duplicata.')
````
